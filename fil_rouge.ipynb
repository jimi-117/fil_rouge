### Now exgtract imdb_id from imdb_link

```python
def get_id(string):
    """
    >>> get_id("https://www.imdb.com/title/tt0111161/")
    'tt0111161'
    """
    words = string.split("/")
    id = [word for word in words if word.startswith("tt")]
    return "".join(id)

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
df_1["id_imdb"] = df_1["imdb_link"].apply(get_id)
df_1.head(5)
```

```python

df_1 = df_1.drop("imdb_link", axis=1)
```

```python
import ast

def list_clean(string):
    """
    >>> list_clean('["a", "b"]')
    ['a', 'b']
    """
    return ast.literal_eval(string)
```

```python
df_1["genre"] = df_1["genre"].apply(list_clean)
df_1["directors"] = df_1["directors"].apply(list_clean)
df_1["actors"] = df_1["actors"].apply(list_clean)
df_1.head(5)
```
